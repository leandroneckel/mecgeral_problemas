# 2.13





Obs.: A menor força necessária P será aquela que fará 90 graus com a força de 50 N (chamada de F na resolução).





```python
f = 50
alpha = 180-90-25
p,r = sp.symbols('p r')
eq1 = sp.Eq(p/sp.sin(25*sp.pi/180),f/sp.sin(alpha*sp.pi/180))
sol1 = sp.solve(eq1,p)
p_value = sp.N(sol1[0])
eq2 = sp.Eq(f/sp.sin(alpha*sp.pi/180),r/sp.sin(90*sp.pi/180))
sol2 = sp.solve(eq2,r)
r_value = sp.N(sol2[0])
print(f'P = {p_value} N, R = {r_value} N')
```

# 2.18







```python
fa = 15
fb = 10
r, alpha, beta = sp.symbols('r alpha beta')
eq1 = sp.Eq(sp.Pow(r,2),sp.Pow(fa,2)+sp.Pow(fb,2)-2*fa*fb*sp.cos(120*sp.pi/180))
sol1 = sp.solve(eq1,r)
r_value = [s.evalf() for s in sol1 if s.is_real and s > 0]
r_value = r_value[0]
eq2 = sp.Eq(r_value/sp.sin(120*sp.pi/180),fa/sp.sin(alpha*sp.pi/180))
sol2 = sp.solve(eq2,alpha)
alpha_value = min(sol2)
eq3 = sp.Eq(fa/sp.sin(alpha_value*sp.pi/180),fb/sp.sin(beta*sp.pi/180))
sol3 = sp.solve(eq3,beta)
beta_value = min(sol3)
print(f'R = {r_value} N, alpha = {alpha_value} graus, beta = {beta_value} graus')
```

Como beta é menor que 40 graus, a resultante realmente aponta para a direita conforme indicado no diagrama de resolução. Então, o ângulo que a resultante faz com a vertical é dado por 40 - beta:

```python
print(40-beta_value)
```

# 2.20





Neste caso, não é mais possível concluir que a resultante é horizontal. Assim:



```python
f = 50
p = 75
r, alpha, beta = sp.symbols('r alpha beta')
eq1 = sp.Eq(sp.Pow(r,2),sp.Pow(f,2)+sp.Pow(p,2)-2*f*p*sp.cos(105*sp.pi/180))
sol1 = sp.solve(eq1,r)
r_value = [s.evalf() for s in sol1 if s.is_real and s > 0]
r_value = r_value[0]
eq2 = sp.Eq(r_value/sp.sin(105*sp.pi/180),f/sp.sin(alpha*sp.pi/180))
sol2 = sp.solve(eq2,alpha)
alpha_value = min(sol2)
eq3 = sp.Eq(f/sp.sin(alpha_value*sp.pi/180),p/sp.sin(beta*sp.pi/180))
sol3 = sp.solve(eq3,beta)
beta_value = min(sol3)
print(f'R = {r_value} N, alpha = {alpha_value} graus, beta = {beta_value} graus')
```

Como alpha é menor que 50 graus, a resultante realmente aponta para baixo da resultante conforme o diagrama de resolução. Assim, a resultante fica orientada para baixo da horizontal sob o ângulo dado por
