```python
import math as mt
import numpy as np
import sympy as sp
```

# 2.7



```python
f = 50
p = 35
ang = 25
alpha, gamma, r = sp.symbols('alpha gamma r')
eq1 = sp.Eq(p/sp.sin(25*sp.pi/180),f/sp.sin(alpha*sp.pi/180))
sol1 = sp.solve(eq1,alpha)
alpha_value = sp.N(min(sol1))
gamma_value = 180-25-alpha_value
eq2 = sp.Eq(f/sp.sin(alpha_value*sp.pi/180),r/sp.sin(gamma_value*sp.pi/180))
sol2 = sp.solve(eq2,r)
r_value = sol2[0]
print(f'alpha = {alpha_value} graus, R = {r_value} N')
```

# 2.8



```python
gamma = 180-35-20
p = 75
r,q = sp.symbols('r q')
eq1 = sp.Eq(r/sp.sin(gamma*sp.pi/180),p/sp.sin(35*sp.pi/180))
sol1 = sp.solve(eq1,r)
r_value = sp.N(sol1[0])
eq2 = sp.Eq(p/sp.sin(35*sp.pi/180),q/sp.sin(20*sp.pi/180))
sol2 = sp.solve(eq2,q)
q_value = sp.N(sol2[0])
print(f'R = { r_value} N, Q = {q_value} N')
```

# 2.9



```python
f = 1600
alpha = 25
gamma = 180-(90-15)-alpha
p,r = sp.symbols('p r')
eq1 = sp.Eq(f/sp.sin(alpha*sp.pi/180),p/sp.sin(75*sp.pi/180))
sol1 = sp.solve(eq1,p)
p_value = sp.N(sol1[0])
eq2 = sp.Eq(p_value/sp.sin(75*sp.pi/180),r/sp.sin(gamma*sp.pi/180))
sol2 = sp.solve(eq2,r)
r_value = sp.N(sol2[0])
print(f'P = {p_value} N, R = {r_value} N')
```

# 2.10






Neste caso, foi utilizada a lei dos cossenos de uma forma menos convencional para esses exerccios: no lugar de deixa a resultante isolada, foi deixada a fora P isolada (eq1). Entretanto, para isso foi necessrio expor no cosseno o ngulo oposto a essa fora (75)

```python
r = 2500
f = 1600
p, alpha = sp.symbols('p alpha')
eq1 = sp.Eq(sp.Pow(p,2),sp.Pow(r,2)+sp.Pow(f,2)-2*r*f*sp.cos(75*sp.pi/180))
sol1 = sp.solve(eq1,p)
sol1 = [sp.N(x) for x in sol1]
sol1 = [s.evalf() for s in sol1 if s.is_real and s > 0]
p_value = sol1[0]
eq2 = sp.Eq(f/sp.sin(alpha*sp.pi/180),p_value/sp.sin(75*sp.pi/180))
sol2 = sp.solve(eq2,alpha)
alpha_value = sp.N(min(sol2))
print(f'P = {p_value} N, alpha = {alpha_value} graus')
```

```python
f = 1890
p = 2220
r,alpha = sp.symbols('r alpha')
eq1 = sp.Eq(f/sp.sin((90-alpha)*sp.pi/180),p/sp.sin(60*sp.pi/180))
sol1 = sp.solve(eq1,alpha)
sol1 = [sp.N(x) for x in sol1]
sol1 = [s.evalf() for s in sol1 if s.is_real and s > 0]
alpha_value = sol1[0]
gamma = 180-(90-alpha_value)-60
eq3 = sp.Eq(p/sp.sin(60*sp.pi/180),r/sp.sin(gamma*sp.pi/180))
sol3 = sp.solve(eq3,r)
r_value = sol3[0]
print(f'alpha = {alpha_value} graus, R = {r_value} N')
```
