```python
import math as mt
import numpy as np
import sympy as sp
```

# 2.35







```python
f1 = 200
f2 = 150
f3 = 100
alpha = 35
ang2 = alpha+30
f1x = f1*mt.cos(alpha*mt.pi/180)
f1y = f1*mt.sin(alpha*mt.pi/180)
f2x = f2*mt.cos(ang2*mt.pi/180)
f2y = f2*mt.sin(ang2*mt.pi/180)
f3x = f3*mt.cos(alpha*mt.pi/180)
f3y = f3*mt.sin(alpha*mt.pi/180)
frx =-f1x+f2x+f3x
fry = -f1y-f2y-f3y
fr = mt.sqrt(mt.pow(frx,2)+mt.pow(fry,2))
angr = mt.atan(abs(fry/frx))*180/mt.pi
print(f'FR = {fr} N, fazendo {angr} com a horizontal')
```
