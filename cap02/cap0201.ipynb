```python
import math as mt
import numpy as np
import sympy as sp
```

```python
p = 75
q = 125
gamma = 180 - 55
r, alpha, beta = sp.symbols('r alpha beta')
eq1 = sp.Eq(sp.Pow(r,2),sp.Pow(p,2)+sp.Pow(q,2)-2*p*q*sp.cos(gamma*sp.pi/180))
sol = sp.solve(eq1,r)
r_value = [s.evalf() for s in sol if s.is_real and s > 0]
r_value = r_value[0]
eq2 = sp.Eq(r_value/sp.sin(gamma*sp.pi/180),p/sp.sin(beta*sp.pi/180))
sol2 = sp.solve(eq2,beta)
beta_value = min(sol2)
eq3 = sp.Eq(r_value/sp.sin(gamma*sp.pi/180),q/sp.sin(alpha*sp.pi/180))
sol3 = sp.solve(eq3,alpha)
alpha_value = min(sol3)
```

Obs.: Não foi utilizada o método do paralelogramo para obtenção da resultante. Foi utilizado o método do polígono através da leis dos cossenos diretamente.

Para a lei dos cossenos, foi isolado o cateto R.

As direções encontradas são as direções entre a resultante e cada uma das forças. Alpha é a orientação de R em relação a P e Beta é a orientação de R em relação a Q.

```python
print(f'R = {r_value},beta = {beta_value}, alpha = {alpha_value}')
```
