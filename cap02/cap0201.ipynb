```python
import math as mt
import numpy as np
import sympy as sp
```

# 2.1

```python
p = 75
q = 125
gamma = 180 - 55
r, alpha, beta = sp.symbols('r alpha beta')
eq1 = sp.Eq(sp.Pow(r,2),sp.Pow(p,2)+sp.Pow(q,2)-2*p*q*sp.cos(gamma*sp.pi/180))
sol = sp.solve(eq1,r)
r_value = [s.evalf() for s in sol if s.is_real and s > 0]
r_value = r_value[0]
eq2 = sp.Eq(r_value/sp.sin(gamma*sp.pi/180),p/sp.sin(beta*sp.pi/180))
sol2 = sp.solve(eq2,beta)
beta_value = min(sol2)
eq3 = sp.Eq(r_value/sp.sin(gamma*sp.pi/180),q/sp.sin(alpha*sp.pi/180))
sol3 = sp.solve(eq3,alpha)
alpha_value = min(sol3)
```

Obs.: No foi utilizada o mtodo do paralelogramo para obteno da resultante. Foi utilizado o mtodo do polgono atravs da leis dos cossenos diretamente.

Para a lei dos cossenos, foi isolado o cateto R.

As direes encontradas so as direes entre a resultante e cada uma das foras. Alpha  a orientao de R em relao a P e Beta  a orientao de R em relao a Q.

```python
print(f'R = {r_value} N,beta = {beta_value} graus, alpha = {alpha_value} graus')
```

# 2.3

```python
tab = 500
tad = 160
alpha = mt.atan(2/2.5)*180/mt.pi
beta = mt.atan(1.5/2.5)*180/mt.pi
gamma = 180 - alpha - beta
r,theta,phi = sp.symbols('r theta phi')
eq1 = sp.Eq(sp.Pow(r,2),sp.Pow(tab,2)+sp.Pow(tad,2)-2*tab*tad*sp.cos(gamma*sp.pi/180))
sol = sp.solve(eq1,r)
r_value = [s.evalf() for s in sol if s.is_real and s > 0]
r_value = r_value[0]
eq2 = sp.Eq(r_value/sp.sin(gamma*mt.pi/180),tab/sp.sin(phi*mt.pi/180))
sol2 = sp.solve(eq2,phi)
phi_value = min(sol2)
eq3 = sp.Eq(r_value/sp.sin(gamma*mt.pi/180),tad/sp.sin(theta*mt.pi/180))
sol3 = sp.solve(eq3,theta)
theta_value = min(sol3)
theta_value
```

Theta  a orientao da TAB em relao a R. Se theta for maior que alpha, isso indica que a resultante aponta para a direita, caso contrrio a resultante aponta para a esquerda

```python
print(f'R = {r_value} N, phi = {phi_value} graus, theta = {theta_value} graus')
```

```python
f1 = 2
f2 = 3
gamma = 80
r, alpha, beta = sp.symbols('r alpha beta')
eq1 = sp.Eq(sp.Pow(r,2),sp.Pow(f1,2)+sp.Pow(f2,2)-2*f1*f2*sp.cos(gamma*sp.pi/180))
sol1 = sp.solve(eq1,r)
r_value = [s.evalf() for s in sol1 if s.is_real and s > 0]
r_value = r_value[0]
eq2 = sp.Eq(r_value/sp.sin(gamma*sp.pi/180),f1/sp.sin(beta*sp.pi/180))
sol2 = sp.solve(eq2,beta)
beta_value = min(sol2)
eq3 = sp.Eq(r_value/sp.sin(gamma*sp.pi/180),f2/sp.sin(alpha*sp.pi/180))
sol3 = sp.solve(eq3,alpha)
alpha_value = min(sol3)
```

```python
print(f'R = {r_value}, alpha = {alpha_value}, beta = {beta_value}')
```
