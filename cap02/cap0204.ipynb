# 2.21





```python
f1 = 800
d1 = mt.sqrt(800*800+600*600)
f1x = f1*800/d1
f1y = f1*600/d1
print(f'A força F1 = 800 N tem F1x = {f1x} N e F1y = {f1y} N')
```

```python
f2 = 424
d2 = mt.sqrt(mt.pow(900,2)+mt.pow(560,2))
f2x = f2*560/d2
f2y = f2*900/d2
print(f'A força F2 = 800 N tem F2x = {f2x} N e F2y = {f2y} N')
```

```python
f3 = 408
d3 = mt.sqrt(mt.pow(900,2)+mt.pow(480,2))
f3x = f3*480/d3
f3y = f3*900/d3
print(f'A força F3 = 800 N tem F3x = {f3x} N e F3y = {f3y} N')
```

# 2.22



```python

forcas = [220,129,227]
ddx = [700,2100,1200]
ddy = [2400,2000,2250]
for i in range(len(forcas)):
    dx = ddx[i]
    dy = ddy[i]
    f = forcas[i]
    dd = mt.sqrt(mt.pow(dx,2)+mt.pow(dy,2))
    fx = f*dx/dd
    fy = f*dy/dd
    print(f'A força de {f} N tem Fx = {fx} N e Fy = {fy} N')
```

# 2.23





```python
forcas = [220,180,260]
ang_x = [90-50,60,25]
# A lista de ângulos acima apresenta o ângulo com o eixo x de cada uma das forças
for i in range(len(forcas)):
    ff = forcas[i]
    ang = ang_x[i]
    fx = ff*mt.cos(ang*mt.pi/180)
    fy = ff*mt.sin(ang*mt.pi/180)
    print(f'A força de {ff} N tem Fx = {fx} N e Fy = {fy} N')
```

# 2.25







```python
px = 1330
p = sp.symbols('p')
eq = sp.Eq(px,p*sp.sin(35*sp.pi/180))
sol = sp.solve(eq,p)
p_value = sp.N(sol[0])
py = p_value*mt.cos(35*mt.pi/180)
print(f'P = {p_value} N, Py = {py} N')
```

```python
f1 = 220
f2 = 129
f3 = 227
d1 = mt.sqrt(mt.pow(700,2)+mt.pow(2400,2))
f1x = f1*700/d1
f1y = f1*2400/d1
d2 = mt.sqrt(mt.pow(2100,2)+mt.pow(2000,2))
f2x = f1*2100/d2
f2y = f1*2000/d2
d3 = mt.sqrt(mt.pow(1200,2)+mt.pow(2250,2))
f3x = f1*1200/d3
f3y = f1*2250/d3
frx = -f1x+f2x+f3x
fry = f1y+f2y-f3y
fr = mt.sqrt(mt.pow(frx,2)+mt.pow(fry,2))
ang = mt.atan(fry/frx)*180/mt.pi
print(f'FR = {fr} N com ângulo com a horizontal de {ang} graus')
```
