# 2.21





```python
f1 = 800
d1 = mt.sqrt(800*800+600*600)
f1x = f1*800/d1
f1y = f1*600/d1
print(f'A força F1 = 800 N tem F1x = {f1x} N e F1y = {f1y} N')
```

```python
f2 = 424
d2 = mt.sqrt(mt.pow(900,2)+mt.pow(560,2))
f2x = f2*560/d2
f2y = f2*900/d2
print(f'A força F2 = 800 N tem F2x = {f2x} N e F2y = {f2y} N')
```

```python
f3 = 408
d3 = mt.sqrt(mt.pow(900,2)+mt.pow(480,2))
f3x = f3*480/d3
f3y = f3*900/d3
print(f'A força F3 = 800 N tem F3x = {f3x} N e F3y = {f3y} N')
```

# 2.22



```python

forcas = [220,129,227]
ddx = [700,2100,1200]
ddy = [2400,2000,2250]
for i in range(len(forcas)):
    dx = ddx[i]
    dy = ddy[i]
    f = forcas[i]
    dd = mt.sqrt(mt.pow(dx,2)+mt.pow(dy,2))
    fx = f*dx/dd
    fy = f*dy/dd
    print(f'A força de {f} N tem Fx = {fx} N e Fy = {fy} N')
```

# 2.23





```python
forcas = [220,180,260]
ang_x = [90-50,60,25]
# A lista de ângulos acima apresenta o ângulo com o eixo x de cada uma das forças
for i in range(len(forcas)):
    ff = forcas[i]
    ang = ang_x[i]
    fx = ff*mt.cos(ang*mt.pi/180)
    fy = ff*mt.sin(ang*mt.pi/180)
    print(f'A força de {ff} N tem Fx = {fx} N e Fy = {fy} N')
```
